```python
# Line plot
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10, 12, 8, 15, 9]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# Bar Plot: Used for comparing categories(not continous data)
import matplotlib.pyplot as plt
categaries = ["Apple", "Banana", "Oranges"]
sales  = [50, 30, 70]
plt.bar(categaries, sales)
plt.title("Fruit Sales")
plt.xlabel("Fruit Type")
plt.ylabel("Quantity sold")
plt.show()
```

```python
import matplotlib.pyplot as plt

height = [150, 160, 165, 170, 175, 180]
weight = [50, 55, 58, 65, 70, 72]

plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (cm)")
plt.show()
```

```python
import matplotlib.pyplot as plt

ages = [22, 25, 22, 30, 28, 35, 30, 30, 40, 42, 50, 55, 55, 55, 60]

plt.hist(ages, bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 16]
y2 = [5, 6, 7, 8, 9]

plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.plot(x, y1)
plt.title("Line Plot")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.bar(x, y2)
plt.title("Bar Chart")

plt.tight_layout()  # avoid overlapping
plt.show()
```
